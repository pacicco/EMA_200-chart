```python
#Download dos dados do Yahoo Finance

#market_data = yf.download(data, one_year_ago, today)
market_data = yf.download('AAPL', start="2000-01-01")

market_data
```

```python


#Cálculo da média móvel exponencial


#today = datetime.datetime.now()
#one_year_ago = today - datetime.timedelta (days = 365)



exponential_moving_average = 200
#exponential_moving_average = 200
EMA_200 = market_data['Close'].ewm(span = exponential_moving_average, adjust = False).mean()

#market_data = yf.download('AAPL', start="2000-01-01")
#market_data['SMA 200'] = ta.sma(data['Close'], 200)


display(EMA_200)
```

```python

#Gráfico plotado com a média móvel


plt.figure(figsize = (16, 8))
plt.plot(market_data['Close'], label = "Apple")
plt.plot(EMA_200, label = "EMA 200")

plt.legend()
```
